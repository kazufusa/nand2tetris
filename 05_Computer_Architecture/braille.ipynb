```python
def drawText(text):
    ttfontname = "/usr/share/fonts/truetype/DejaVuSans.ttf"
    fontsize = 60
    canvasSize    = (40, 80)
    backgroundRGB = (255, 255, 255)
    textRGB       = (0, 0, 0)

    img  = PIL.Image.new('RGB', canvasSize, backgroundRGB)
    draw = PIL.ImageDraw.Draw(img)

    font = PIL.ImageFont.truetype(ttfontname, fontsize)
    draw.text((0,0), text, fill=textRGB, font=font)
    return img
```

```python
xs = [30, 15]
ys = [15, 30 , 45, 60]

def tobyte(img):
    a = np.array(img)
    return [ "1" if (a[y, x]==0).all() else "0" for y in ys for x in xs]


for text in texts:
    b = tobyte(drawText(text))
    print(f"{{{', '.join(b)}}}: '{text}',")
```
